Підключаємо необхідні бібліотеки

```python
import pandas as pd
```

**Вхідні дані**

```python
df = pd.read_csv('table_10.csv',header=None)
table = df.values
```

```python
data = pd.read_csv('prob_10.csv',header=None)
prob = data.values

P_M = [0 for i in range(0,20)]
for i in range(0,20):
  P_M[i] = prob[0][i]

P_k = [0 for i in range(0,20)]
for i in range(0,20):
  P_k[i] = prob[1][i]
```

Знайдемо розподіл P(C)

```python
P_C = []

for C in range(0,20):
  t = 0
  for i in range(0,20):
    for j in range(0,20):
      if (table[i][j] == C):
        t = t + P_M[j] * P_k[i]
  P_C.append(t)

print(P_C)
```

Знайдемо розподіл P(M,C)

```python
P_MC = [[0 for c in range(0,20)] for r in range(0,20)]

for C in range(0,20):
  for M in range(0,20):
    t = 0
    for i in range(0,20):
      if (table[i][M] == C):
        t = t + P_M[M] * P_k[i]
    P_MC[M][C] = t

print(P_MC)
```

Обчислимо P(M|C)

```python
P_M_C = [[0 for c in range(0,20)] for r in range(0,20)]

for i in range(0,20):
  for j in range(0,20):
    P_M_C[i][j] = P_MC[i][j] / P_C[j]

print(P_M_C)
```

Знайдемо оптимальну детерміністичну вирішувальну функцію

```python
P_C_M = [[0 for c in range(0,20)] for r in range(0,20)]

for i in range(0,20):
  for j in range(0,20):
    P_C_M[i][j] = P_M_C[j][i]
```

```python
delta_d = [[0 for c in range(0,20)] for r in range(0,20)]

for i in range(0,20):
  for j in range(0,20):
    if (P_C_M[i].index(max(P_C_M[i])) == j):
      delta_d[i][j] = 1

print(delta_d)
```

Знайдемо функцію втрат детерміністичної функції

```python
loss_d = 0

for i in range(0,20):
  loss_d = loss_d + P_C[i] * (1 - max(P_C_M[i]))

print(loss_d)
```

Знайдемо оптимальну стохастичну вирішувальну функцію

```python
delta_s = [[0 for c in range(0,20)] for r in range(0,20)]
t = [0 for i in range(20)]

for i in range(0,20):
  m = max(P_C_M[i])
  for j in range(0,20):
    if (P_C_M[i][j] == m):
      delta_s[i][j] = max(P_C_M[i])
      t[i] = t[i] + P_M[j]

  for j in range(0,20):
    if (delta_s[i][j] != 0):
      delta_s[i][j] = P_M[j] / t[i]

print(delta_s)
```

```python
loss_s = 0

for i in range(0,20):
  for j in range(0,20):
    loss_s = loss_s + P_MC[j][i]*(1 - delta_s[i][j])

print(loss_s)
```
